```python
from pca import PCA

import matplotlib.pyplot as plt
from sklearn.datasets import load_iris,load_breast_cancer
```

<h2> Iris Dataset </h2>

```python
dataset = load_iris()

X = dataset.data
y = dataset.target
```

```python
pca = PCA(k=2)

X_transformed = pca.fit_transform(X)
```

```python
plt.scatter(X_transformed[:,0],X_transformed[:,1],c=y)
```

<h2> Breast Cancer Dataset </h2>

```python
pca = PCA(k=2)

X_transformed = pca.fit_transform(X)
```
