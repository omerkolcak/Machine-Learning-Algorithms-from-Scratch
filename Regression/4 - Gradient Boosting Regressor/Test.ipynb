```python
from GradientBoostingRegressor import GradientBoostingRegressor

import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("dataset.csv")
df.head()
```

```python
x = df.drop("Price",axis=1)
y = df['Price'].values
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
y_train = pd.DataFrame(y_train,columns=['Price'],index=x_train.index)
```

```python
gb = GradientBoostingRegressor(max_depth=3,n_estimators=10)
```

```python
gb.fit(x_train,y_train,learning_rate=0.25)
```

```python
test_preds = gb.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_percentage_error
```

```python
mean_absolute_percentage_error(test_preds,y_test)
```
